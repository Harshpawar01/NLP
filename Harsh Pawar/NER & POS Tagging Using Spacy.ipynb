Ques.1 Create a Doc object from the file peterrabbit.txt

```python
import spacy
nlp = spacy.load("en_core_web_sm")
with open("peterrabbit.txt") as file:
    text = file.read()

doc = nlp(text)
```

 Ques.2 For every token in the third sentence, print the token text, the POS tag, the fine-grained TAG tag, and the description of the fine-grained tag.


```python
sentences = list(doc.sents)
third_sentence = sentences[2]
for token in third_sentence:
    print(f"Token: {token.text}, POS: {token.pos_}, TAG: {token.tag_}, Description: {spacy.explain(token.tag_)}")
```

Ques.3 Provide a frequency list of POS tags from the entire document

```python
from collections import Counter
pos_tags = [token.pos_ for token in doc]
pos_freq = Counter(pos_tags)
print(pos_freq)
```

Ques.4 CHALLENGE: What percentage of tokens are nouns?

```python
noun_count = sum(1 for token in doc if token.pos_ == "NOUN")
total_tokens = len(doc)

# Calculate percentage
noun_percentage = (noun_count / total_tokens) * 100
print(f"Percentage of nouns: {noun_percentage:.2f}%")
```

Ques.5  Display the Dependency Parse for the third sentence.

```python
from spacy import displacy

displacy.render(third_sentence, style="dep",options={'distance':110})
```

Ques.6 Show the first two named entities from Beatrix Potter's The Tale of Peter Rabbit


```python
entities = list(doc.ents)
for ent in entities[:2]:
    print(ent.text, ent.label_)
```

Ques.7 How many sentences are contained in The Tale of Peter Rabbit?


```python
num_sentences = len(list(doc.sents))
print(f"Number of sentences: {num_sentences}")
```

Ques.8 CHALLENGE: How many sentences contain named entities?


```python
sentences_with_ents = sum(1 for sent in doc.sents if any(token.ent_type_ for token in sent))
print(f"Sentences with named entities: {sentences_with_ents}")
```
